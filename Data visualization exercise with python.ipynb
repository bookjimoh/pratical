```python
x = np.linspace(0,5,11)
y = x**2

plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.xlabel('My x axis')
plt.ylabel('My y axis')
plt.title('This is my left title')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.xlabel('My x axis')
plt.ylabel('My y axis')
plt.title('This is my right title')
```

## Object Oriented Method

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y,'r')
axes1.set_xlabel('This is x large')
axes1.set_ylabel('This is large y')
axes1.set_title('This is the title (big)')
axes2.plot(x,y,'b')
axes2.set_xlabel('This is x small')
axes2.set_ylabel('This is snall y')
axes2.set_title('This is small title (big)')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('This is a left title')
axes[0].set_xlabel('This is x on the left')
axes[0].set_ylabel('This is x on the left')
axes[1].plot(x,y)
axes[1].set_title('This is a right title')
axes[1].set_xlabel('This is x on the right')
axes[1].set_ylabel('This is x on the right')

plt.tight_layout()
```

```python
covid_data = pd.read_csv('covid_data.csv')
```

```python
covid_data
```

```python
covid_data.head(2)
```

```python
country = 'China'
country_data = covid_data[covid_data['location']==country]

fig,axes = plt.subplots(figsize = (12,6))
axes.plot(country_data['date'], country_data['Total_cases'])
axes.set_xlabel('Dates')
axes.set_ylabel('Total Number of cases')
axes.set_title(f'This is the total number of cases for {country}')

fig,axes = plt.subplots(figsize = (12,6))
axes.plot(country_data['date'], country_data['Total_deaths'])
axes.set_xlabel('Dates')
axes.set_ylabel('Total Number of deaths')
axes.set_title(f'This is the total number of deaths for {country}')
```

```python
date = "2021-01-01"
date_data = covid_data[covid_data['date']==date]

fig,axes = plt.subplots(figsize = (12,6))

axes.scatter(date_data['Total_cases'], date_data['Total_deaths'])
axes.set_xlabel('This is total cases')
axes.set_ylabel('This is total deaths')
axes.set_title('Total Cases vs. total deadths')
```
